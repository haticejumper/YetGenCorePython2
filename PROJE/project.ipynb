```python
# Dizi film 
# 1- Api yada elle veri toplanacak
# 2-random film - dizi önerme
# 3- türe göre sınıflandırma ve öneri -H
# 3-filtreleme -C
# 3-oyuncuya yönetmene göre filtreleme yapılabilir 
# 5-ımdb puanına göre ekleme çıkarma 
# 6- düzenleme 
# 8-benzeyen dizi önerme (kategoriler) 
# Şunları da beğenebilirsin köşesi 
# 9-favorileri duzenle -C
# puana gore siralama -H
# data frame ekleme


class Movie:
    def __init__(self, name, director, genre, year, rating, log):
        self.name = name
        self.director = director
        self.genre = genre
        self.year = year
        self.rating = rating
        self.log = log

class Series:
    def __init__(self, name, director, genre, year, rating, log, seasons):
        self.name = name
        self.director = director
        self.genre = genre
        self.year = year
        self.rating = rating
        self.log = log
        self.seasons = seasons

movies = []
tvSeries = []
favorites = []



# add movie
def add_movie():
    print("Enter movie details: ")
    name = input("Name: ")
    director = input("Director: ")
    genre = input("Genre: ")
    year = input("Year: ")
    rating = input("Rating: ")
    log = input("Log: ")

    movie = Movie(name, director, genre, year, rating, log)
    movies.append(movie)

    isFavorite = input("Do you want to add this movie into favorites (y/n): ")
    if isFavorite == "y":
        favorites.append(movie)

    print(f"{name} is added!")

# remove movie
def remove_movie():
    name = input("Enter movie name to remove: ")
    for movie in movies:
        if movie.name == name:
            movies.remove(movie)
            print(f"{name} has been removed.")
            break
    else:
        print(f"{name} not found.")

# list movies
def list_movies():
    if len(movies) == 0:
        print("Movies are not found!")

    for movie in movies:
        print(f"{movie.name} ({movie.year}) - {movie.genre} - Directed by {movie.director} - Rating: {movie.rating}")

# add series 
def add_series():
    print("Enter series details: ")
    name = input("Name: ")
    director = input("Director: ")
    genre = input("Genre: ")
    year = input("Year: ")
    rating = input("Rating: ")
    log = input("Log: ")
    seasons = input("Seasons: ")

    series = Series(name, director, genre, year, rating, log)
    tvSeries.append(series)

    isFavorite = input("Do you want to add this series into favorites (y/n): ")
    if isFavorite == "y":
        favorites.append(series)

    print(f"{name} is added!")

#remove series
def remove_series():
    name = input("Enter movie name to remove: ")
    for series in tvSeries:
        if series.name == name:
            tvSeries.remove(series)
            print(f"{name} has been removed.")
            break
    else:
        print(f"{name} not found.")

# list series
def list_series():
    if len(tvSeries) == 0:
        print("Series are not found!")

    for series in tvSeries:
        print(f"{series.name} ({series.year}) - {series.genre} - Directed by {series.director} - Rating: {series.rating}")

'''import random
random_suggestion= random.choice(Movie)
print(random_suggestion)
'''


print("""
Select an operation:
0. Exit 
1. Add movie
2. Remove movie
3. List movies
4. Add Series 
5. Remove Series
6. List Series
""")

while True:
    choice = input("Enter your choice: ")

    if choice == "1":
        add_movie()
    elif choice == "2":
        remove_movie()
    elif choice == "3":
        list_movies()
    elif choice == "4":
        add_series()
    elif choice == "5":
        remove_series()
    elif choice == "6":
        list_series()
    elif choice == "0":
        break
    else:
        print("Invalid choice. Try again.")
```
