# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
def kaydet():
    isim = input("Ürün ismini giriniz: ")
    fiyat = input("Ürün fiyatini giriniz: ")
    with open("ürünler.txt" ) as f:
        f.write(f"{isim},{fiyat}\n")
    print("kaydedildi")
````

# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
def degiştir(eski_kelime , yeni_kelime):
    with open("ürünler.txt", "r+") as f:
        text = f.read()
        new_text = text.replace(eski_kelime , yeni_kelime)
        f.seek(0)
        f.write(new_text)
        f.truncate()
    print("Kelimeler başariyla değiştirildi.")
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def get_info(self):
        return f"{self.make} {self.model} ({self.year})"


class Bus(Vehicle):
    def __init__(self, make, model, year, capacity):
        super().__init__(make, model, year)
        self.capacity = capacity

    def get_info(self):
        return f"{super().get_info()} - Capacity: {self.capacity}"
````
